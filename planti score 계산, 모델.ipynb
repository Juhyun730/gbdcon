# Planty_score : 당산비*0.3 +  당도 *0.7

```python
x=raw[['6월 평균 기온', '8월 평균 기온', '5월 강수량', '10월 일교차', '3월 일교차', '9월 강수량', '8월 일교차']]
```

```python
def getValue(value):
    if value <= q[0.25]:
        return 1
    elif value <= q[0.5]:
        return 2
    elif value <= q[0.75]:
        return 3
    else:
        return 4
    
q = raw["PLANTY_SCORE"].quantile([0.25, 0.5, 0.75, 1])

for i in range(37):
    raw.loc[i, ['new']] = getValue(raw.iloc[i]['PLANTY_SCORE'])

#######'new'값이 y값으로 들어감
```

```python
raw
```

```python
y=raw[['new']]
```

```python
from imblearn.over_sampling import SMOTE  #데이터의 불균형한 라밸링개수를 맞추기 위함
X_resampled, y_resampled = SMOTE(random_state=42).fit_resample(x,y)
y_resampled.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X_resampled, y_resampled, test_size=0.20, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier(max_depth=10, min_samples_leaf=3)
tree.fit(X_train, y_train)
pred_train=tree.predict(X_test)
tree.score(X_test, y_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier  #모델 돌리기

rand_clf = RandomForestClassifier(bootstrap=True,
                                  max_depth = 10,
                                  min_samples_leaf = 5,
                                  min_samples_split = 3,
                                  n_estimators = 200)
rand_clf.fit(X_train, y_train.values.ravel())


pred=rand_clf.predict(X_test)
accuracy = accuracy_score(y_test, pred)
accuracy
```
