# 면적당 생산량

```python
import pandas as pd
import numpy as np

fruit_output = pd.read_csv("dataset/yield_final_final.csv")
fruit_output['new'] = fruit_output['생산량']/fruit_output['면적'] 
```

사과만 뽑아보기

```python
fruit_output = fruit_output[fruit_output['작물']== '사과']
fruit_output
```

단위면적당 생산량 높은순으로

```python
grouped = fruit_output['new'].groupby(fruit_output['지역'])
grouped.mean().sort_values(ascending = False)
```

```python
import requests

url = 'http://apis.data.go.kr/1360000/AsosDalyInfoService/getWthrDataList'
params ={'serviceKey' : '70tdOp3V5FJxQdk+vCg/SB6ISNTaKcdZA5LuFMnfUyWoCp5kw1kGOx97IwXru3PLRmPcs+436viY2Z6aJBuVwg==',
         'pageNo' : '1', 'numOfRows' : '365', 'dataType' : 'json', 'dataCd' : 'ASOS', 'dateCd' : 'DAY', 'startDt' : '20150101', 'endDt' : '20181231', 'stnIds' : '108' }

response = requests.get(url, params=params)
data=response.content
data=json.loads(data)
data
```

```python
df=json_normalize(data['response']['body']['items']['item'])
df=df[['stnNm','tm','avgTa','minTa','maxTa','sumRn','sumSsHr']] #지역이름,시간, 평균온도,최소온도,최고온도,일강수량합,합계일조시간
df['sumRn'] = df['sumRn'].replace(r'^\s*$', 0, regex=True)#강수량 빈칸인 날짜는 0으로 넣기
#월별로 평균 내보기
df['년월']=df['tm'].str[:7]
df=df.drop(['tm'],axis=1)
df[['평균온도','최소온도','최고온도','일강수량합','합계일조시간']]= df[['avgTa','minTa','maxTa','sumRn','sumSsHr']].apply(pd.to_numeric)
df=df.drop(['avgTa','minTa','maxTa','sumRn','sumSsHr'],axis=1)
df=df.groupby(['년월'], as_index=False).mean()
df
```
